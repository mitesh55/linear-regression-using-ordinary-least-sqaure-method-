# **Linear Regression using Ordinary Least Square Method :**

linear regression meaning in simple word is to find best fit line (y = mx+c) which gives least error to estimate y value based on x data input. 



 

# First, import the libraries :

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

# **Load dataset :**

```python
data = pd.read_csv('../input/salary-datacsv/Salary_Data.csv')
data.head()
```

#  **After loading dataset we need to split the data into feature and label :**

  here we want to predict salary based on years of experience, so y will be label and x will be feature

```python
x = data.iloc[:,0]
y = data.iloc[:,1]
```

#  After spliting data visulize data using matplotlib :

```python
plt.scatter(x, y)
```

# Using Seaborn :

```python
sns.scatterplot(x, y, data= data)
```

# Inference :

from above fig. we can say that there is a linear and positive relation exist between x & y
to find relation we use correlation coefficient method.


# Correlation Coefficient (r):
  
  Degree of association between x & y represents by r (-1,1) to check relation between them. 
  
  if r = -1 : strong relation but in negative direction
  if r = 1 : storng relation in positive direction
  if r = 0 : then no relation between x & y so we can't predict y value based on x data.
  if r = 0-0.5 : relation is weak between x & y in positive direction
  if r = 0.5 - 0.8 : relation is moderate between x & y 




```python
n= len(x)
xy = x*y
Exy = sum(xy)
Ex = sum(x)
Ey = sum(y)
x_sq = x*x
y_sq = y*y
Ex_sq = sum(x_sq)
Ey_sq = sum(y_sq)

r1 = (n*Exy - Ex*Ey)
r2 = ((n*Ex_sq-Ex**2)*(n*Ey_sq-Ey**2))**(1/2)

r = r1/r2
r
```

# Pair plot :
pairplot gives graphical representation of relation between each pair

```python
sns.pairplot(data)
```

so from above information we can conclude that if we get r value more than 0.8 or less than - 0.8, that means 
we can continue to predict y based on x data and here we get r = 0.9782 (>0.8)


# Ordinary Least Square Method :

now we'll use ordinary least sqaure method to predict y (salary)
experience = m ( slop), intercept = c 
salary = experience * (x) + intercept
E is sigma(sum of data) 

              
 Experience = E(Yi-mean(y))(Xi-mean(x)) / E(Xi-mean(x))^2
               
    
 intercept = Y - experience * (X) 


```python
x_mean = x.mean()
y_mean = y.mean()

Yi_y_mean = y-y_mean
Xi_x_mean = x-x_mean

Xi_x_mean_sq = (x-x_mean)**2

exp_nominator = sum(Yi_y_mean * Xi_x_mean)

exp_denominator = sum(Xi_x_mean_sq)

experience = exp_nominator/exp_denominator

intercept = y_mean - (experience * x_mean)
```
